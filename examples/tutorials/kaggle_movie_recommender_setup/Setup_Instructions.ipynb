### Convert the `ipynb` file and create the requirements

What is `nbconvert` and why do we need it? 
- `pipreqs` only works on `.py` files. We have to convert the `.ipynb` file to a `.py` file

Setup: I already created a `./reqs` directory, but you can also create one using the UI or `mkdir ./reqs`

```python
!jupyter nbconvert --output-dir="./reqs" --to script ./getting-started-with-a-movie-recommendation-system.ipynb
!cd ./reqs && pipreqs
```
